##### Analysis results using PCMDI Metrics Package.
The PCMDI metrics package is used to objectively compare results from climate models with observations and each other using well-established statistical tests. Summary statistics are produced in the context of all model simulations contributed to CMIP5 and earlier CMIP phases.

The metrics used to generate portrait plots in this notebook can be found here- ftp://nomads.gfdl.noaa.gov/users/Ming.Zhao/AM4Documentation/GFDL-AM4.0/JAMES/supplementary/PMP/metrics_latest_v1p1p1/share/CMIP_metrics_results/CMIP5/

The metrics computed for GFDL-AM4/AMIP can be found here-
ftp://nomads.gfdl.noaa.gov/users/Ming.Zhao/AM4Documentation/GFDL-AM4.0/JAMES/supplementary/PMP/pcmdi_metrics_v1p1p1/metrics/

Please note that the above two are input to this notebook.

This notebook provides an interactive set of commands to generate Portrait plot for the PCMDI metrics generated.
